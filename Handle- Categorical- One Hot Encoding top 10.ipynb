```python
# One Hot Encoding - top 10
```

```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_columns",None)
```

```python
df=pd.read_csv('mercedes.csv', usecols=['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df.X0.unique()
```

```python
# each column have too many values for one hot encoding, so we will do hot encoding for top 10 features only for each column
# Lets handle X0 column first
df.shape
pd.get_dummies(df)
```

```python
# here 7 columns created new 170 columns so 
# find unique values in each column

for i in df.columns:
    print(i, len(df[i].unique()) )
```

```python
# wrt to columns, we will do one hot encoding for 10 most freq variables only
top_10=df.X0.value_counts(ascending=False).head(10).index
top_10
```

```python
top_10=list(top_10)
top_10
```

```python
def one_hot(df,variable,top_10):
    for i in top_10:
        df[variable+'_'+i]=np.where(df[variable]==i,1,0)
```

```python
# One hot for X0

one_hot(df,'X0',top_10)
```

```python
df
```

```python
# do same for X1

top_10=[x for x in df['X1'].value_counts().head(10).index]
top_10
```

```python
# ond hot for X1

one_hot(df,'X1',top_10)
```
