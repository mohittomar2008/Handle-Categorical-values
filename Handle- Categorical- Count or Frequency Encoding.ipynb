# Count and Frequency Encoding

```python
# import data from the link

import pandas as pd
import numpy as np
```

```python
train_set= pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None,index_col=None)
```

```python
train_set
```

```python
# choose selected categorical columns only
columns=[1,3,5,6,7,8,9,13]
```

```python
train_set=train_set[columns]
```

```python
# set columns name
train_set.columns=['Employment','Degree','Marrige_status',"Designation",'Relation','Race','Gender','Country']
```

```python
train_set
```

```python
# We are going to find unique values in each column and then

train_set['Employment'].value_counts().to_dict()
```

```python
# we will replace all categorical features with the frequency of its.
```

```python
#lets see how many unique labels in each columns]
for i in train_set.columns:
    print("column" ,i,"have", len(train_set[i].unique()), "unique labels" )
```

```python
# function to replace categorical feature with its frequency

def ok(df, variable):
    map_freq=train_set[variable].value_counts().to_dict()
    train_set[variable]=train_set[variable].map(map_freq)
```

```python
# Need to run this for all columns

for i in train_set.columns:
    ok(train_set,i)
```
